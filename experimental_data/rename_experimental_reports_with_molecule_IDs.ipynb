## Rename pKa experiment reports with SAMPL6 molecule IDs

In this noteboook, I will add SAMPL6 molecule IDs as prefixes to Sirius T3 experiment report files.

```python
import os
import pandas as pd
```

```python
path_to_experimental_data_table = "pKa_replicate_experimental_results.csv"
input_path_to_reports_with_exp_ID = "./experiment_reports_with_experimental_molecule_ID/"
output_path_to_reports_with_mol_ID ="./experiment_reports_with_molecule_ID/"

# copy the report files in experimental ID directory to molecule ID directory
! cp $input_path_to_reports_with_exp_ID*.pdf $output_path_to_reports_with_mol_ID
```

```python
df_exp = pd.read_csv(path_to_experimental_data_table)
df_exp
```

```python
# iterate over report names to save the new report files with molecule ID prefixes

for i, row in enumerate(df_exp.iterrows()):
    old_report_name = row[1]["Experiment Report"]
    molecule_ID = row[1]["Molecule ID"]
    new_report_name = molecule_ID + "_" + old_report_name + ".pdf"
    old_report_name = old_report_name + ".pdf"
    
    old_file = os.path.join(output_path_to_reports_with_mol_ID, old_report_name)
    new_file = os.path.join(output_path_to_reports_with_mol_ID, new_report_name)
    #print(old_file)
    #print(new_file)
    #print()
    
    # rename report file
    os.rename(old_file, new_file)
```
